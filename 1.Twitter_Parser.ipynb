### Парсинг

```python
'''По тексту'''
text = 'Подари жизнь' #здесь пишем нужный текст
config = twint.Config()
config.Search = text
config.Store_csv = True
config.Limit = 10000
config.Output = f'{text}.csv'
twint.run.Search(config)
```

```python
'''По юзернейму'''
username = 'podarizhizn' #здесь пишем нужное имя
config = twint.Config()
config.Username = 'podarizhizn'
config.Store_csv = True
#config.Limit = 10000
config.Output = f'from_user-{username}.csv'
twint.run.Search(config)
```

### Преобразование

```python
tweets = pd.read_csv(f'{text}.csv')
tweets = tweets[['id','date','username','tweet','replies_count','retweets_count','likes_count','hashtags']]
tweets.head(10)
```

```python
tweets.to_csv(f'{text}.csv', index = False)
```

```python
tweets = pd.read_csv(f'from_user-{username}.csv')
tweets = tweets[['id','date','username','tweet','replies_count','retweets_count','likes_count','hashtags']]
tweets.head(10)
```
